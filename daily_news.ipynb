```python
# Define keywords to search for and set the minimum keyword frequency threshold
keywords = ["Apple", "AAPL"]
min_frequency = 5  # Customize this threshold based on how relevant you want the content to be

# Function to count keyword occurrences in title and content
def count_keyword_frequency(row, keywords):
    count = 0
    for keyword in keywords:
        count += row["title"].count(keyword) + row["content"].count(keyword)
    return count

# Apply the keyword counting function to each row
df2["keyword_frequency"] = df2.apply(lambda row: count_keyword_frequency(row, keywords), axis=1)
# Filter rows with keyword frequency meeting the minimum threshold
df2_filtered = df2[df2["keyword_frequency"] >= min_frequency]

# Display the filtered DataFrame
print("Filtered DataFrame with relevant news about Apple:")
```

```python
from llama import LlamaLLM
import re

NAME = 'Apple'
llm = LlamaLLM()
```

```python
import spacy

# Load the spaCy English language model for sentence segmentation
nlp = spacy.load("en_core_web_sm")

# Function to filter and extract Apple-related sentences
def extract_news_helper(content):
    # Define Apple-related keywords
    keywords = {"apple", "aapl", "ios", "macos", "ipad", "iphone", "apple tv"}
    content = re.sub(r'^\s*\n', '.', content, flags=re.MULTILINE)
    # Use spaCy to segment text into sentences
    doc = nlp(content)
    
    # Extract sentences containing Apple-related keywords
    related_sentences = [
        sent.text for sent in doc.sents
        if any(keyword.lower() in sent.text.lower() for keyword in keywords)
    ]
    
    # Join the relevant sentences into a single string
    related_content = " ".join(related_sentences)
    return related_content

def extract_relevant_news(row):

    related_content = extract_news_helper(row['content'])
    row['related_content'] = related_content
    
    return row

df2_filtered = df2_filtered.apply(extract_relevant_news, axis=1)
```

```python
# feed ke LLM sentiment analysis --> 
df2_filtered.to_csv('./data/aapl_filtered.csv', index=False)
```
